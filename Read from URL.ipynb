```python
import requests

# URL of the webpage
url = "https://pudhari.news/vishwasanchar/ai-impersonates-brad-pitt-scams-woman-7-crore"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Specify the file name to save the HTML content
    file_name = "downloaded_source.html"
    
    # Open the file in write mode and save the content
    with open(file_name, "w", encoding="utf-8") as file:
        file.write(response.text)

    print(f"HTML source code has been saved as '{file_name}'")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import requests

# URL of the webpage
url = "https://pudhari.news/vishwasanchar/ai-impersonates-brad-pitt-scams-woman-7-crore"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the text content of the HTML
    html_content = response.text

    # Find the content between 'articleBody":' and 'लोकल'
    start_index = html_content.find('articleBody":')  # Locate the start of 'articleBody":'
    end_index = html_content.find('लोकल', start_index)  # Locate the position of 'लोकल'

    if start_index != -1 and end_index != -1:
        # Extract content between 'articleBody":' and 'लोकल'
        extracted_content = html_content[start_index + len('articleBody":'): end_index].strip()
        print("Extracted Content:\n", extracted_content)
    else:
        print("Unable to find the specified content.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import requests

# URL of the webpage
url = "https://www.esakal.com/premium-article/saptahik/pegasus-exploring-technology-in-journalism-loren-and-sandrick-book-marathi-review-sk88"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the raw HTML content
    html_content = response.text
    
    # Find the starting and ending indices for extracting the content
    start_index = html_content.find('"articleBody":')  # Locate the start of 'articleBody":'
    end_index = html_content.find('सकाळ', start_index)  # Locate the end string 'सकाळ'

    if start_index != -1 and end_index != -1:
        # Extract content between '"articleBody":' and 'सकाळ'
        extracted_content = html_content[start_index + len('"articleBody":'): end_index].strip()

        # Print the extracted content
        print("Extracted Content:\n", extracted_content)
    else:
        print("Unable to find the specified content.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import requests
import tkinter as tk
from tkinter import messagebox

# Function to fetch and extract content from the URL
def fetch_and_extract():
    url = url_entry.get()  # Get the URL from the textbox

    if not url:  # Check if the URL is empty
        messagebox.showwarning("Input Error", "Please enter a valid URL.")
        return

    try:
        # Send HTTP GET request to the URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            html_content = response.text

            # Find the starting and ending indices for extracting the content
            start_index = html_content.find('"articleBody":')  # Locate the start of 'articleBody":'
            end_index = html_content.find('सकाळ', start_index)  # Locate the end string 'सकाळ'

            if start_index != -1 and end_index != -1:
                # Extract content between '"articleBody":' and 'सकाळ'
                extracted_content = html_content[start_index + len('"articleBody":'): end_index].strip()

                # Display the extracted content in the text area
                result_text.delete(1.0, tk.END)  # Clear any existing text
                result_text.insert(tk.END, extracted_content)  # Insert the extracted content

            else:
                messagebox.showerror("Extraction Error", "Unable to find the specified content.")
        else:
            messagebox.showerror("Request Error", f"Failed to retrieve the webpage. Status code: {response.status_code}")

    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {str(e)}")

# Setting up the Tkinter window
window = tk.Tk()
window.title("Extract Content from URL")
window.geometry("1200x1000")  # Set the size of the window

# Label for URL input
url_label = tk.Label(window, text="Enter URL:")
url_label.pack(pady=10)

# Textbox for URL input
url_entry = tk.Entry(window, width=50)
url_entry.pack(pady=5)

# Button to trigger content extraction
read_button = tk.Button(window, text="Read", command=fetch_and_extract)
read_button.pack(pady=10)

# Text area to display the extracted content
font_style = ("Helvetica", 12)  # Increase font size by 5 (from default 8 to 12)
result_text = tk.Text(window, width=220, height=100, font=font_style)
result_text.pack(pady=10)

# Run the Tkinter event loop
window.mainloop()
```
